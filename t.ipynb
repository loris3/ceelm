```python
import torch
from collections import defaultdict
import pandas as pd
from tqdm import tqdm


class DummyModel:
    def __init__(self, val_grad_dict, tr_grad_dict):
        self.val_grad_dict = val_grad_dict
        self.tr_grad_dict = tr_grad_dict
        self.n_train = len(tr_grad_dict)
        self.hvp_dict = {}
        self.IF_dict = {}

    def get_gradient(self, tokenized_train_dataset, train_dataset_name, train_dataset_split, batch_ids):
        return [self.tr_grad_dict[tr_id] for tr_id in batch_ids]

 
    def compute_hvp_batch(
        self,
        batch_size=265,
        val_batch_size=64,
        lambda_const_param=10,
        inner_batch_size=2
    ):
        hvp_proposed_dict = defaultdict(dict)
        n_train = len(self.tr_grad_dict)
        val_ids = list(self.val_grad_dict.keys())
        print("val_ids", len(val_ids))
    
        sample_val_id = next(iter(self.val_grad_dict))
        weight_names = list(self.val_grad_dict[sample_val_id].keys())
        n_weights = len(weight_names)

        print("Computing lambda_const...")
        
        sum_sq = torch.zeros(n_weights)
        for i in tqdm(range(0, n_train, batch_size), desc="Compute lambda_const"):
            batch_ids = list(range(i, min(i + batch_size, n_train)))
            batch_grads_list = self.get_gradient(None, None, None, batch_ids)
            for grads_dict in batch_grads_list:
                grads = torch.stack([grads_dict[w].view(-1) for w in weight_names])
                sum_sq += grads.pow(2).mean(dim=1)
        lambda_const = (sum_sq / n_train) / lambda_const_param


        print("Computing HVP (single pass over train data per val batch)...")
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

        for vb_start in tqdm(range(0, len(val_ids), val_batch_size), desc="Validation batches"):
            vb_end = min(vb_start + val_batch_size, len(val_ids))
            val_batch = val_ids[vb_start:vb_end]

       
            val_grads_batch = torch.stack([
                torch.stack([self.val_grad_dict[v][w].view(-1) for w in weight_names])
                for v in val_batch
            ])
            hvp_batch = torch.zeros_like(val_grads_batch) 


            val_grads_batch = val_grads_batch.to(device)
            lambda_const = lambda_const.to(device)
            hvp_batch = hvp_batch.to(device)

            for i in tqdm(range(0, n_train, batch_size),desc="Inner train data loop"):
                batch_ids = list(range(i, min(i + batch_size, n_train)))
                batch_grads_list = self.get_gradient(None, None, None, batch_ids)

                batch_grads = torch.stack([
                    torch.stack([batch_grads_list[j][w].view(-1) for w in weight_names])
                    for j in range(len(batch_ids))
                ]).to(device)   

                

                for k in range(0, batch_grads.size(0), inner_batch_size):
                    sub_batch_grads = batch_grads[k:k+inner_batch_size] 
                    sub_dot = (sub_batch_grads.unsqueeze(0) * val_grads_batch.unsqueeze(1)).sum(dim=3)
                    sub_denom = lambda_const.view(1, 1, -1) + sub_batch_grads.pow(2).sum(dim=2).unsqueeze(0)
                    sub_C_tmp = sub_dot / sub_denom

                    hvp_batch += (
                        (val_grads_batch.unsqueeze(1) - sub_C_tmp.unsqueeze(3) * sub_batch_grads.unsqueeze(0))
                        / (n_train * lambda_const.view(1, -1, 1))
                    ).sum(dim=1)


                del batch_grads
                torch.cuda.empty_cache()
     
            for v_idx, val_id in enumerate(val_batch):
                for w_idx, w in enumerate(weight_names):
                    hvp_proposed_dict[val_id][w] = hvp_batch[v_idx, w_idx].float()

        return hvp_proposed_dict

    def compute_hvp_simple(self, lambda_const_param=10):
        hvp_proposed_dict = defaultdict(dict)
        for val_id in tqdm(self.val_grad_dict.keys(), desc="Computing hvp_simple"):
            for weight_name in self.val_grad_dict[val_id]:
                S = torch.zeros(len(self.tr_grad_dict))
                for tr_id in self.tr_grad_dict:
                    tmp_grad = self.tr_grad_dict[tr_id][weight_name]
                    S[tr_id] = torch.mean(tmp_grad ** 2)
                lambda_const = torch.mean(S) / lambda_const_param

                hvp = torch.zeros_like(self.val_grad_dict[val_id][weight_name])
                for tr_id in self.tr_grad_dict:
                    tmp_grad = self.tr_grad_dict[tr_id][weight_name]
                    C_tmp = torch.sum(self.val_grad_dict[val_id][weight_name] * tmp_grad) / (lambda_const + torch.sum(tmp_grad ** 2))
                    hvp += (self.val_grad_dict[val_id][weight_name] - C_tmp * tmp_grad) / (self.n_train * lambda_const)
                hvp_proposed_dict[val_id][weight_name] = hvp
        return hvp_proposed_dict


def compute_IF(self):
    for method_name in self.hvp_dict:
        print("Computing IF for method: ", method_name)
        if_tmp_dict = defaultdict(dict)
        for tr_id in tqdm(self.tr_grad_dict, desc="Naive Dot products",  total=len(self.tr_grad_dict)):
            for val_id in self.val_grad_dict:
                if_tmp_value = 0
                for weight_name in self.val_grad_dict[0]:
                    if_tmp_value += torch.sum(self.hvp_dict[method_name][val_id][weight_name]*self.tr_grad_dict[tr_id][weight_name])
                if_tmp_dict[tr_id][val_id]=-if_tmp_value

        self.IF_dict[method_name] = pd.DataFrame(if_tmp_dict, dtype=float)   


def compute_IF_batched(self, batch_size: int = 128):
        tr_ids = list(self.tr_grad_dict.keys())
        val_ids = list(self.val_grad_dict.keys())
        weight_names = list(self.val_grad_dict[val_ids[0]].keys())

        n_train = len(tr_ids)
        n_val = len(val_ids)
        if_matrix = torch.zeros(n_train, n_val, dtype=torch.float32)


        hvp_flat_dict = {
            val_id: torch.cat([self.hvp_dict["proposed"][val_id][w].view(-1) for w in weight_names]).float()
            for val_id in val_ids
        }

  
        for i in tqdm(range(0, n_train, batch_size), desc="Train batches"):
            batch_ids = tr_ids[i:i + batch_size]
            batch_grads_list = self.get_gradient(
                None, None, None, batch_ids
            )
            batch_grads = torch.stack([
                torch.cat([grad[w].view(-1) for w in weight_names]).float()
                for grad in batch_grads_list
            ])

          
            for v_idx, val_id in enumerate(val_ids):
                hvp_flat = hvp_flat_dict[val_id]

                if_batch = -torch.matmul(batch_grads, hvp_flat)
                if_matrix[i:i + len(batch_ids), v_idx] = if_batch.detach()

            del batch_grads, batch_grads_list
            torch.cuda.empty_cache()

        self.IF_dict["proposed"] = pd.DataFrame(
            if_matrix, index=tr_ids, columns=val_ids, dtype=float
        ).T


torch.manual_seed(0)
tr_grad_dict = {i: {f'w{i}': torch.randn(16, 16,dtype=torch.float32) for i in range(4)} for i in range(1110)}
val_grad_dict ={i: {f'w{i}': torch.randn(16, 16,dtype=torch.float32) for i in range(4)} for i in range(111)}


model = DummyModel(val_grad_dict, tr_grad_dict)
hvp_batch = model.compute_hvp_batch()
hvp_simple = model.compute_hvp_simple()


model.hvp_dict = {'proposed': hvp_simple}


compute_IF(model)
df_orig = model.IF_dict['proposed']

model.IF_dict = {}  # reset
compute_IF_batched(model, batch_size=2)
df_batched = model.IF_dict['proposed']

print("Original IF:\n", df_orig)
print("\nBatched IF:\n", df_batched)

all_close = torch.allclose(torch.tensor(df_orig.values),
                           torch.tensor(df_batched.values),
                           atol=1e-3)
print("\nAre IFs equivalent?", all_close)
```
